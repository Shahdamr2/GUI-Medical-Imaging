```python
from tkinter import *
from PIL import ImageTk, Image, ImageFilter
from tkinter import filedialog
import numpy as np
import cv2
import PIL.Image, PIL.ImageTk
import tkinter.font as font




pro = Tk()
pro.title("GUI Medical Imaging")
pro.geometry("1500x1500")
pro.resizable(True, True)
pro.config(background='gray')
frame_tool = Frame(pro, bg='#663399', height=790)
frame_tool.pack(fill='x')
myFont = font.Font(family="Elephant", size=12, weight="bold")



def identify_image(): 
    global canvas, IMG, image, image_tk
    canvas = Canvas(pro, width=400, height=400)
    canvas.place(x=500, y=120)
    IMG = filedialog.askopenfilename()
    image = Image.open(IMG)
    image_tk = ImageTk.PhotoImage(image)
    image = image.resize((400, 400), Image.ANTIALIAS)
    image_tk = ImageTk.PhotoImage(image)
    canvas.create_image(0, 0, anchor=NW, image=image_tk)

    
def adaptiveThresholdMean():
    img = cv2.imread(IMG, 0)
    img = cv2.resize(img, (400,400), interpolation=cv2.INTER_AREA)
    _,th1=cv2.threshold(img,100,255,cv2.THRESH_BINARY)
    th6=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(th6))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    root.mainloop()
    
    
def adaptiveThresholdGaussian():
    img = cv2.imread(IMG, 0)
    img = cv2.resize(img, (400,400), interpolation=cv2.INTER_AREA)
    _,th1=cv2.threshold(img,100,255,cv2.THRESH_BINARY)
    th8=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(th8))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    root.mainloop()    
    
    
def otsu_thresholding():
    img = cv2.imread(IMG, 0)
    img = Image.fromarray((img))
    img = np.array(img, dtype=np.uint8)
    img = cv2.GaussianBlur(img, (5, 5), 0)
    ret1, th1 = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
    ret2, th2 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    th1 = cv2.resize(th1, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(th1))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()
    
    
    
def Sobel_x():
    img = cv2.imread(IMG)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
    sobel_x=np.uint8(np.absolute(sobel_x))
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(sobel_x))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop() 
    
    
def Sobel_y():
    img = cv2.imread(IMG)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    sobel_y=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
    sobel_y=np.uint8(np.absolute(sobel_y))
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(sobel_y))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop() 
    
    
def Canny():
    img = cv2.imread(IMG)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
    sobel_x=np.uint8(np.absolute(sobel_x))
    sobel_y=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
    sobel_y=np.uint8(np.absolute(sobel_y))
    sobel_combined=cv2.bitwise_or(sobel_x,sobel_y)
    canny_img=cv2.Canny(sobel_combined,600,300)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(canny_img))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()     
    
    
    
def THRESH_BINARY():
    img = cv2.imread(IMG, 0)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    ret,th1=cv2.threshold(img,100,255,cv2.THRESH_BINARY)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(th1))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    root.mainloop()
    
    
def THRESH_BINARY_INV():
    img = cv2.imread(IMG, 0)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    ret,th2=cv2.threshold(img,100,255,cv2.THRESH_BINARY_INV)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(th2))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    root.mainloop()    
    

    
    
def THRESH_TRUNC():
    img = cv2.imread(IMG, 0)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    ret,th3=cv2.threshold(img,100,255,cv2.THRESH_TRUNC)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(th3))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    root.mainloop()
    
    
def THRESH_TOZERO():
    img = cv2.imread(IMG, 0)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    ret,th4=cv2.threshold(img,100,255,cv2.THRESH_TOZERO)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(th4))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    root.mainloop()
    
    
def THRESH_TOZERO_INV():
    img = cv2.imread(IMG, 0)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    ret,th5=cv2.threshold(img,100,255,cv2.THRESH_TOZERO_INV)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(th5))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    root.mainloop()    
    
def contors():
    img = cv2.imread(IMG)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret,img_threshold = cv2.threshold(img_gray, 15, 255,0)
    contours, hierarchy = cv2.findContours(img_threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    print("number of contours = " + str(len(contours)))
    print(contours[0])
    cv2.drawContours(img,contours,11,(255, 0, 0),3)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(img))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()    
    
def LevelSet():
    def compute_gradient_magnitude(image, mask):
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        grad_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
        grad_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
        gradient_magnitude = np.sqrt(grad_x*2 + grad_y*2)
        gradient_magnitude[~mask] = 0  # Set gradient magnitude outside mask to zero
        return gradient_magnitude
    
    def initialize_level_set(image_shape, init_mask):
        level_set = np.zeros(image_shape[:2], dtype=np.float32)
        level_set[init_mask] = -1
        level_set[~init_mask] = 1
        return level_set
    
    def evolve_level_set(image, level_set, mask, iterations=100, alpha=0.5, beta=0.02):
        for i in range(iterations):
            gradient_magnitude = compute_gradient_magnitude(image, mask)
            laplacian = cv2.Laplacian(gradient_magnitude, cv2.CV_64F)
        
            gradient = np.gradient(level_set)
            gradient_norm = np.sqrt(gradient[0]*2 + gradient[1]*2)
        
            force = alpha * laplacian - beta * gradient_norm
            level_set += force
        
            print(f"Iteration {i+1}: Min value: {np.min(level_set)}, Max value: {np.max(level_set)}")
        return level_set

# Function to apply level set result to image
    def apply_level_set(image, level_set):
        result = image.copy()
        level_set_normalized = cv2.normalize(level_set, None, 0, 255, cv2.NORM_MINMAX)
        result[:, :, 0] = level_set_normalized.astype(np.uint8)  # Apply to blue channel
        result[:, :, 1] = level_set_normalized.astype(np.uint8)  # Apply to green channel
        result[:, :, 2] = level_set_normalized.astype(np.uint8)  # Apply to red channel
        return result

# Read input image
    image = cv2.imread('Contour.jpeg')
    init_mask = np.zeros(image.shape[:2], dtype=bool)
    init_mask[300:500, 300:500] = True  # Initialize mask for the segmented region

# Initialize level set function
    level_set = initialize_level_set(image.shape, init_mask)

# Evolve level set
    evolved_level_set = evolve_level_set(image, level_set, init_mask)

# Apply level set result to image
    result = apply_level_set(image, evolved_level_set)

    img = cv2.resize(result, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(result))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    root.mainloop()
    
    
    

def connected_component_label():
    img = cv2.imread(IMG, 0)
    img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]
    num_labels, labels = cv2.connectedComponents(img)
    
    label_hue = np.uint8(179 * labels / np.max(labels))
    blank_ch = 255 * np.ones_like(label_hue)
    labeled_img = cv2.merge([label_hue, blank_ch, blank_ch])

    labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)
    labeled_img[label_hue == 0] = 0
    labeled_img = cv2.resize(labeled_img, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(labeled_img))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()
    
    
    
    
def watershed():
    from skimage import measure, color, io
    img = cv2.imread(IMG)
    img = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    cells=img[:,:,0] 
    ret1, thresh = cv2.threshold(cells, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

    kernel = np.ones((3,3),np.uint8)
    opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 1)
    from skimage.segmentation import clear_border
    opening = clear_border(opening) 
    
    sure_bg = cv2.dilate(opening,kernel,iterations=1)
 
    dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
    ret2, sure_fg = cv2.threshold(dist_transform,0.3*dist_transform.max(),255,0)
    # Finding unknown region
    sure_fg = np.uint8(sure_fg)  #Convert to uint8 from float
    unknown = cv2.subtract(sure_bg,sure_fg)

    # Marker labelling
    ret3, markers = cv2.connectedComponents(sure_fg)
    # Add one to all labels so that sure background is not 0, but 1
    markers = markers+10
    # Now, mark the region of unknown with zero
    markers[unknown==255] = 0
    markers = cv2.watershed(img,markers)
    img[markers == -1] = [255,255,255] 
    img2 = color.label2rgb(markers, bg_label=0)
    img2 = cv2.resize(img, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(img2))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()
    

    
def chaincode():
    from matplotlib import pyplot as plt
    image = cv2.imread(IMG, 0)
    rows, cols = image.shape
    result = np.zeros_like(image)
    for x in range(rows):
        for y in range(cols):
            if image[x, y] >= 70:
                result[x, y] = 0
            else:
                result[x, y] = 1
    
    for i, row in enumerate(result):
        for j, value in enumerate(row):
            if value == 1:
                start_point = (i, j)
                #            print(start_point, value)
                break
        else:
            continue
        break

    directions = [0, 1, 2,
                  7, 3,
                  6, 5, 4]
    dir2idx = dict(zip(directions, range(len(directions))))
    # print(dir2idx)
    change_j = [-1, 0, 1,  # x or columns
                -1, 1,
                -1, 0, 1]

    change_i = [-1, -1, -1,  # y or rows
                0, 0,
                1, 1, 1]

    border = []
    chain = []

    curr_point = start_point
    for direction in directions:
        idx = dir2idx[direction]
        print(idx)
        new_point = (start_point[0] + change_i[idx], start_point[1] + change_j[idx])
        print(new_point)
        if result[new_point] != 0:  # if is ROI
            border.append(new_point)
            chain.append(direction)
            curr_point = new_point
            break
    count = 0
    while curr_point != start_point:
        # figure direction to start search
        b_direction = (direction + 5) % 8
        dirs_1 = range(b_direction, 8)
        dirs_2 = range(0, b_direction)
        dirs = []
        dirs.extend(dirs_1)
        dirs.extend(dirs_2)
        for direction in dirs:
            idx = dir2idx[direction]
            new_point = (curr_point[0] + change_i[idx], curr_point[1] + change_j[idx])
            if result[new_point] != 0:  # if is ROI
                border.append(new_point)
                chain.append(direction)
                curr_point = new_point
                break
        if count == 1000: break
        count += 1
    print(count)
    print(chain)
    plt.imshow(image, cmap='Greys')
    plt.plot([i[1] for i in border], [i[0] for i in border])
    plt.imshow(image, cmap='Greys')
    image = cv2.resize(image, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(image))
    plt.show()
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()
    
    
def k_means():
    img = cv2.imread(IMG)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img_val=img.reshape((-1,3))
    img_val=np.float32(img_val)
    criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,100,0.2)
    k=3
    _,labels,(centers)=cv2.kmeans(img_val,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
    centers=np.uint8(centers)
    labels=labels.flatten()
    seg_img=centers[labels.flatten()]
    seg_img=seg_img.reshape(img.shape)
    seg_img = cv2.resize(seg_img, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(seg_img))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()
    
    
def Region_Growing():
    def get8n(x, y, shape):
        out = []
        maxx = shape[1]-1
        maxy = shape[0]-1

    #top left
        outx = min(max(x-1,0),maxx)
        outy = min(max(y-1,0),maxy)
        out.append((outx,outy))

    #top center
        outx = x
        outy = min(max(y-1,0),maxy)
        out.append((outx,outy))

    #top right
        outx = min(max(x+1,0),maxx)
        outy = min(max(y-1,0),maxy)
        out.append((outx,outy))

    #left
        outx = min(max(x-1,0),maxx)
        outy = y
        out.append((outx,outy))

    #right
        outx = min(max(x+1,0),maxx)
        outy = y
        out.append((outx,outy))

    #bottom left
        outx = min(max(x-1,0),maxx)
        outy = min(max(y+1,0),maxy)
        out.append((outx,outy))

    #bottom center
        outx = x
        outy = min(max(y+1,0),maxy)
        out.append((outx,outy))

    #bottom right
        outx = min(max(x+1,0),maxx)
        outy = min(max(y+1,0),maxy)
        out.append((outx,outy))

        return out

    def region_growing(img, seed):
        list = []
        outimg = np.zeros_like(img)
        list.append((seed[0], seed[1]))
        processed = []
        while(len(list) > 0):
            pix = list[0]
            outimg[pix[0], pix[1]] = 255
            for coord in get8n(pix[0], pix[1], img.shape):
                if img[coord[0], coord[1]] != 0:
                    outimg[coord[0], coord[1]] = 255
                    if not coord in processed:
                        list.append(coord)
                processed.append(coord)
            list.pop(0)
        return outimg

    def on_mouse(event, x, y, flags, params):
        if event == cv2.EVENT_LBUTTONDOWN:
            print( 'Seed: ' + str(x) + ', ' + str(y), img[y,x])
            clicks.append((y,x))

    clicks = []
    image = cv2.imread(IMG, 0)
    ret, img = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY)
    cv2.namedWindow('Input')
    cv2.setMouseCallback('Input', on_mouse, 0, )
    cv2.imshow('Input', image)
    cv2.waitKey()
    seed = clicks[-1]
    out = region_growing(img, seed)
    img = cv2.resize(out, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(img))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()    
    
    
def MeanShift():
    from sklearn.cluster import MeanShift, estimate_bandwidth
    originImg = cv2.imread(IMG)
    originShape = originImg.shape
    flatImg = np.reshape(originImg, [-1, 3])
    bandwidth = estimate_bandwidth(flatImg, quantile=0.1, n_samples=100)
    ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
    ms.fit(flatImg)
    labels = ms.labels_
    # print(labels)
    cluster_centers = ms.cluster_centers_
    # with np.printoptions(threshold=np.inf):
    #    print(cluster_centers)
    labels_unique = np.unique(labels)
    n_clusters_ = len(labels_unique)
    # print("number of estimated clusters : %d" % n_clusters_)
    segmentedImg = cluster_centers[np.reshape(labels, originShape[:2])]
    segmentedImg = np.uint8(segmentedImg)
    segmentedImg = cv2.resize(segmentedImg, (400, 400), interpolation=cv2.INTER_AREA)
    image_tk = PIL.ImageTk.PhotoImage(image=PIL.Image.fromarray(segmentedImg))
    canvas.create_image(0, 0, anchor=NW, image=image_tk)
    pro.mainloop()

    
    
    

p1 = Button(pro, text="Exit", fg='white', bg='#000000', command=pro.destroy)
p1.place(x=700, y=620, height=50, width=70)

p2 = Button(pro, text="Upload", fg='black', bg='white', command=identify_image,
            activebackground="#F9B7FF")
p2.place(x=600, y=620, height=50, width=80)

p20 = Button(pro, text="k_means", fg='black', bg='#E6E6FA', command=k_means)
p20.place(x=240, y=295, height=50, width=120)

p21 = Button(pro, text="MeanShift", fg='black', bg='#E6E6FA', command=MeanShift)
p21.place(x=240, y=420, height=50, width=120)

B27 = Button(pro, text="contors", fg='black', bg='#E6E6FA', command=contors)
B27.place(x=250, y=600, height=50, width=150)

p21 = Button(pro, text="Sobel_x", fg='black', bg='#E6E6FA', command=Sobel_x)
p21.place(x=1140, y=420, height=50, width=120)

p28 = Button(pro, text="Sobel_y", fg='black', bg='#E6E6FA', command=Sobel_y)
p28.place(x=1140, y=480, height=50, width=120)

p28 = Button(pro, text="Canny", fg='black', bg='#E6E6FA', command=Canny)
p28.place(x=1140, y=550, height=50, width=120)


p38 = Button(pro, text="chaincode", fg='black', bg='#E6E6FA', command=chaincode)
p38.place(x=1000, y=20, height=50, width=150)


B3 = Button(pro, text="connected_component_L", fg='black', bg='#E6E6FA', 
             command=connected_component_label)
B3.place(x=1000, y=150, height=50, width=220)

B12 = Button(pro, text="THRESH_BINARY" ,  fg='black', bg='#E6E6FA',  command= THRESH_BINARY)
B12.place(x=10, y=150, height=50, width=120)
B13 = Button(pro, text="THRESH_BINARY_INV", fg='black', bg='#E6E6FA', command=THRESH_BINARY_INV)
B13.place(x=10, y=225, height=50, width=120)
B17 =Button(pro, text="THRESH_TRUNC",  fg='black', bg='#E6E6FA', command=THRESH_TRUNC)
B17.place(x=10, y=360, height=50, width=120)
B22 = Button(pro, text="THRESH_TOZERO",  fg='black', bg='#E6E6FA', command=THRESH_TOZERO)
B22.place(x=10, y=295, height=50, width=100)
B37 =Button(pro, text="THRESH_TOZERO_INV", fg='black', bg='#E6E6FA', command=THRESH_TOZERO_INV)
B37.place(x=10, y=640, height=50, width=120)

B25 = Button(pro, text="otsu_thresholding",  fg='black', bg='#E6E6FA', command=otsu_thresholding)
B25.place(x=10, y=560, height=50, width=160)

B4 = Button(pro, text="adaptiveThresholdMean", fg='black', bg='#E6E6FA', command=adaptiveThresholdMean)
B4.place(x=10, y=420, height=50, width=140)

B5 = Button(pro, text="adaptiveThresholdGaussian", fg='black', bg='#E6E6FA', command=adaptiveThresholdGaussian)
B5.place(x=10, y=500, height=50, width=140)

B19 = Button(pro, text="Region_Growing",  fg='black', bg='#E6E6FA', command=Region_Growing)
B19.place(x=1040, y=260, height=50, width=140)

B5 = Button(pro, text="LevelSet", fg='black', bg='#E6E6FA',  command=LevelSet)
B5.place(x=240, y=360, height=50, width=120)

B8 = Button(pro, text="watershed",  fg='black', bg='#E6E6FA', command=watershed)
B8.place(x=1000, y=90, height=50, width=150)



pro.mainloop()
```
